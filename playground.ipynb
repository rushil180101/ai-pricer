```python
import feedparser

feed_url = "https://bargainbabe.com/amazon-deals/feed/"

d = feedparser.parse(feed_url)

entries = d.entries
```

```python
import requests
from bs4 import BeautifulSoup

with open("temp_logs.log", "w", encoding="utf-8") as log_file:

    for i in range(5):
        print(entries[i].title)
        print(entries[i].summary)
        print(entries[i].link)

        response = requests.get(entries[i].link, timeout=10)
        text = response.text

        soup = BeautifulSoup(text, "html.parser")
        content = soup.find(attrs={"class": "entry-content"})
        desc = content.get_text(strip=True)

        print(desc)

        log_file.write(f"{desc}\n\n")

        # soup = BeautifulSoup(entries[i].summary, "html.parser")
        # summary = soup.find("p").get_text(strip=True)
        # print(summary)
        print("\n\n")
```
