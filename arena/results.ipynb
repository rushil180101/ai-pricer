```python
import matplotlib.pyplot as plt

data = {
    "Random pricer": 451.40,
    "Avg. pricer": 43.72,
    "Linear regression": 48.10,
    "Random forest": 32.40,
    "Support vector regression": 31.15,
    "Elasticnet regression": 41.32,
    "Vanilla neural network": 44.23,
    "Frontier llm (deepseek-r1-0528) - 10 datapoints": 11.05,
    "Open source base model (Qwen/Qwen3-0.6B)": 34.94,
    "Fine tuned open source model (meta-llama-3.2-1B)": 18.77,
}

data = dict(sorted(data.items(), key=lambda item: item[1], reverse=True))

names = list(data.keys())
values = list(data.values())

plt.figure(figsize=(15, 7))
bars = plt.bar(names, values)
plt.xlabel("Models")
plt.ylabel("Error ($)")
plt.title(
    "Prediction errors representing performance of different models (350 datapoints)"
)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

# Add values for readability
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2, height, f"{height}", ha="center", va="bottom"
    )

plt.savefig("model_comparison_results.png", bbox_inches="tight")
plt.show()
plt.close()
```
